## An example with `scikit-learn`

We will treat an extremely simple toy problem, we will try to model the relationship between $X$ and $Y$ with

$$ Y = 2 \times X + 1 $$

To complicate things a bit, we will add a gaussian noise to the output data (at first)




We start by loading the necessary modules




````python
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
````

````python
init_notebook_mode(connected=True)
import numpy as np
np.random.seed(10)
X = np.random.random(10)


Y = 2*X+1+0.2*np.random.randn(10)
````

To use the plotly module correctly we also need to initialize each cell where we are going to plot with the following Python function 'configure_plotly_browser_state()'

Now we create the matrix of observations 

````python
M = np.ones([len(X),2])
M[:,1] = X
````

````python
M
````

and we solve the least squares problem

$$ min_{b, a} ||MA-Y||_2^2$$

where $A = [b,a]$ and the models writes 
$$ y = ax+b $$

````python
b, a = np.linalg.lstsq(M,Y)[0]
````

````python
print(f"a : {a}, b : {b}")
````

*Note*: We use here the `lstsq` method of `numpy` which is precisely made for least squares problems. It is a `python` method which actually calls stable and proven numerical routines (here the SVD of the `M` matrix) from the LAPACK library (in C or Fortran).

We now plot the solution. To visualize it easily, we create a sampling of the segment $[0,1]$ with `np.linspace` and we compute our estimated linear model on this sampling.


````python
configure_plotly_browser_state()
x_gr = np.linspace(0,1,3)
y_gr = a*x_gr+b
trace = Scatter(
    x=X,
    y=Y,
    mode='markers+text',
    marker=dict(size=13),
    text=[f"Point {i+1}" for i in range(10)],
    textposition="top center",
    name="Sample Data",
)
trace2 = Scatter(
    x=x_gr,
    y=y_gr,
    mode='lines',
    line=dict(width=4),
    name="Ordinary Least Squares"
)
fig = dict(data=[trace,trace2], layout={})
fig['layout']['xaxis'] = dict(title='Variable X')
fig['layout']['yaxis'] = dict(title='Variable Y')

iplot(fig)
````

On rejoue ici l'exemple du cours o on dtermine  l'aide d'une base de test le meilleur degr de rgression en visualisant le dilemne biais-variance. Cela nous permet d'illustrer le calcul de l'erreur MSE et les `pipeline`de `sklearn`

On va utiliser dans la suite la classe `Pipeline`qui permet de crer un objet combinant un preprocessing et un modle (ou plusieurs) de rgression ou de classification. Cet objet hrite des mthodes `.fit` et `.predict` et eut tre appel directement sur les entres. Par exemple, si on veut rejouer l'exemple prcdent (rgression de degr 2)

````python
configure_plotly_browser_state()


from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

degree = 2

model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
model.fit(X.reshape(-1, 1), Y)

y_pred = model.predict(X_test.reshape(-1, 1))

MSE = np.mean((y_pred-y_test)**2)
print(f"Test error : {MSE}")
````

On peut maintenant trs facilement boucler sur le degr et calculer l'erreur de test pour chaque degr de $1$  $15$ par exemple.

````python
configure_plotly_browser_state()

deg_min = 1
deg_max = 12

TEST_ERR = []
TRAIN_ERR = []
for d in range(deg_min, deg_max):
  degree = d
  model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
  model.fit(X.reshape(-1, 1), Y)
  y_pred = model.predict(X_test.reshape(-1, 1))
  y_train = model.predict(X.reshape(-1, 1))
  TEST_ERR.append(np.mean((y_pred-y_test)**2))
  TRAIN_ERR.append(np.mean((y_train-Y)**2))

trace = Scatter(
    x=np.linspace(deg_min, deg_max, deg_max-deg_min+1),
    y=TEST_ERR,
    mode='markers+lines',
    marker=dict(size=13,color='green'),
    line=dict(width=4, dash='dash'),
    name="Test Error",
)

trace2 = Scatter(
    x=np.linspace(deg_min, deg_max, deg_max-deg_min+1),
    y=TRAIN_ERR,
    mode='markers+lines',
    marker=dict(size=13,color='#636EFA'),
    line=dict(width=4, dash='dash'),
    name="Train Error",
)

fig = dict(data=[trace, trace2], layout={})
fig['layout']['xaxis'] = dict(title='Degree of polynomial regression')
fig['layout']['yaxis'] = dict(title='MSE')

iplot(fig)
````

##### Calcul des coefficients de la rgression Ridge direct

La formule vue en formation est la suivante :

$$ \alpha_{Ridge}^{\lambda} = (X^T X + \lambda I)^{-1}X^TY $$

comme dit en formation, on vite en gnral d'utiliser explicitement l'inverse d'une matrice comme cette opration est coteuse et est souvent mal conditionne numriquement. En pratique, il vaut mieux chercher $ \alpha_{Ridge}^{\lambda} $ comme solution du systme suivant :
$$ (X^T X + \lambda I)\alpha_{Ridge}^{\lambda} = X^TY $$

La matrice $ (X^T X + \lambda I) $ tant symtrique et a priori bien conditionne (grce  $\lambda$) on peut utiliser une procdure type Choleski pour rsoudre ce systme. On commence par assembler la matrice des observations $X$ (qu'on appelle $M$ dans le code pour viter de la confondre avec les entres) 


````python
M = np.ones([len(X),3])
M[:,1] = X
M[:,2] = X**2
# calcul de la matrice Ridge
lamb = 0.1
alpha_ridge = np.linalg.solve(M.transpose() @ M + lamb*np.eye(3), M.transpose() @ Y )
c_ridge10, b_ridge10, a_ridge10 = alpha_ridge
print(f"a_ridge10 : {a_ridge10}, b_ridge10 : {b_ridge10}, c_ridge10 : {c_ridge10}")
````

On peut comparer avec `sklearn`

````python
from sklearn.linear_model import Ridge 
from sklearn.preprocessing import PolynomialFeatures 
pf = PolynomialFeatures(degree =2)
x_ = pf.fit_transform(X.reshape(-1, 1))
rlr = Ridge(alpha=0.1, fit_intercept=False)
rlr.fit(x_,Y)
alpha_ridgesk = rlr.coef_
c_ridge10sk, b_ridge10sk, a_ridge10sk = alpha_ridgesk

print(f"a_ridge10sk : {a_ridge10}, b_ridge10sk : {b_ridge10}, c_ridge10sk : {c_ridge10}")
````

##### Evolution des coefficients de la Ridge Regression en fonction de $\lambda$

````python
from sklearn.linear_model import Ridge 

degree = 2

pf = PolynomialFeatures(degree =degree)
x_ = pf.fit_transform(X.reshape(-1, 1))
linreg = lm.LinearRegression(fit_intercept=False)

lamb_vec = np.logspace(-6,3,50)
lamb_vec = np.insert(lamb_vec, 0, 0, axis=0)
coef_ridge = []
for lamb in lamb_vec:
  if lamb == 0:
    linreg.fit(x_, Y.reshape(-1, 1))
    coef_ridge.append(linreg.coef_[0])
  else:
    rlr = Ridge(alpha=lamb, fit_intercept=False)
    rlr.fit(x_,Y)
    coef_ridge.append(rlr.coef_)
````

#### Solution

##### Cration des chantillons

````python
import numpy as np
np.random.seed(42)
N = 50
X = np.random.random(N)
Y = -2*X**2+X-1+0.1*np.random.randn(50)
print(f"Generated {N} samples")

X_test = np.random.random(200)
y_test = -2*X_test**2+X_test-1+0.1*np.random.randn(200)
````

##### Calcul des coefficients de la rgression LASSO avec `scikit-learn`



Comme indiqu dans la formation, il n'y a pas de formule explicite pour le calcul des coefficients d'une rgression LASSO. Le calcul effectif repose sur la rsolution d'un problme d'optimisation essentiellement la minimisation de la distance quadratique sous des contraintes linaires. Il existe de nombreux algorithmes pour le faire inspirs des mthodes de points intrieurs, de simplex ou encore ddis. Dans le cas de `scikit-learn`, la classe `Lasso` implmente un algorithme de descente par coordonnes (voir la partie sur les solveurs pour la classification linaire).

```
class sklearn.linear_model.Lasso(alpha=1.0, *, fit_intercept=True, normalize=False, precompute=False, copy_X=True, max_iter=1000, tol=0.0001, warm_start=False, positive=False, random_state=None, selection='cyclic')
```
Enfin, cette classe rsout le problme d'optimisation suivant (lgrement diffrent de celui vu en formation)

$$ \frac{1}{2n_{samples}}||y-X\alpha||_2^2+\lambda||\alpha||_1 $$


````python
from sklearn.linear_model import Lasso 
from sklearn.preprocessing import PolynomialFeatures 

lamb = 0.001
pf = PolynomialFeatures(degree =2)
x_ = pf.fit_transform(X.reshape(-1, 1))
llr = Lasso(alpha=lamb, fit_intercept=False)
llr.fit(x_,Y)
alpha_lassosk = llr.coef_
c_lasso, b_lasso, a_lasso = alpha_lassosk

print(f"a_lasso : {a_lasso}, b_lasso : {b_lasso}, c_lasso : {c_lasso}")
````

On trace ensuite, comme pour la rgression Ridge, l'erreur de test et l'erreur d'apprentissage pour dterminer le meilleur degr de rgression en fonction de la valeur de $\lambda$

On utilise la fonction de l'article cit plus haut

````python
import numpy.linalg as la
def tls(X,y):
    
    if X.ndim is 1: 
        n = 1 # the number of variable of X
        X = X.reshape(len(X),1)
    else:
        n = np.array(X).shape[1] 
    
    Z = np.vstack((X.T,y)).T
    U, s, Vt = la.svd(Z, full_matrices=True)
    
    V = Vt.T
    Vxy = V[:n,n:]
    Vyy = V[n:,n:]
    a_tls = - Vxy  / Vyy # total least squares soln
    
    Xtyt = - Z.dot(V[:,n:]).dot(V[:,n:].T)
    Xt = Xtyt[:,:n] # X error
    y_tls = (X+Xt).dot(a_tls)
    fro_norm = la.norm(Xtyt, 'fro')
    
    return y_tls, X+Xt, a_tls, fro_norm
````

````python
M_init = np.ones([len(X_init),2])
M_init[:,1] = X_init
y_tls,X_tls,coeff_tls,fro_norm = tls(M_init, Y_real)
````

## Gnralits sur la rgression logistique en `sklearn` 

Le module `linear_model` de `sklearn` comprend la classe `LogisticRegression`  qui implmente l'entranement d'un classifieur linaire par rgression logistique. Comme on l'a vu en formation, l'entranement consiste  minimiser la quantit suivante 
$$ \sum_{i=1} \log(e^{-y_i(X_i^Tw+b)}+1) $$
sur les paramtres $w$ et $b$. Contrairement  la rgression logistique souvent implmente dans les logiciels statistiques qui n'introduit pas de terme de rgularisation, `sklearn` a fait le choix de l'implmenter par dfaut. Ainsi, le problme d'optimisation rsolu par `sklearn` est 

$$ \min_{w,b} \frac{1}{2}||w||_2 + C \sum_{i=1} \log(e^{-y_i(X_i^Tw+b)}+1) $$

Par dfaut, `sklearn` implmente donc la rgression ridge mais on peut trs bien utiliser la rgularisation LASSO, sous la forme suivante :

$$ \min_{w,b} ||w||_1 + C \sum_{i=1} \log(e^{-y_i(X_i^Tw+b)}+1) $$

Et enfin on peut trs bien combiner les deux sous la forme de l'_Elastic Net_


$$ \min_{w,b} \rho ||w|+ \frac{1-\rho}{2}||w||_2 + C \sum_{i=1} \log(e^{-y_i(X_i^Tw+b)}+1) $$

Dans tous les cas, augmenter la valeur de $C$ diminue l'impact de la pnalisation. Pour retrouver le cas non rgularis (moins stable numriquement et risque accru d'overfitting), il faut donner une trs grande valeur  ce paramtre. Cette formulation est largement inspire par la librairie crite en C [liblinear](https://www.csie.ntu.edu.tw/~cjlin/liblinear/), librairie historique notamment pour l'apprentissage de modle linaire avec beaucoup d'examples.

## Solveurs et capacits

Lors de la formation, nous avons vu que le problme d'optimisation  rsoudre pour entraner une rgression logistique n'avait
pas de solution simple crite sous une forme analytique ou d'algbre linaire. Il faut donc utiliser des algorithmes d'optimisation numrique. Le plus classique dans le cas de la rgression logistique tant l'IRLS (_Iterative Reweighted Least Squares_) qui n'est autre que la mthode de Newton crite dans le cas spcifique de la rgression logistique. Cette mthode revient  rsoudre  chaque itration de Newton un systme aux moindres carrs pondrs. De trs nombreuses alternatives ont t dveloppes pour pallier les difficults de convergence de IRLS dans certains cas :


*   pnalisation LASSO
*   grande taille (beaucoup d'examples)

Typiquement, on trouve parmi les mthodes dveloppes 



*   Des algorithmes classiques et tout-terrains de type quasi-Newton (`BFGS` ou limited memory `L-BFGS`) souvent efficaces en grande taille sur des objectifs drivables qui peuvent tre parallliss efficacement.
*   les algorithmes de descentes par coordonnes (_coordinate descent_) adaptes  la rgularisation LASSO et  la grande taille car se paralllise aisment, voir [ici](https://www.cs.ubc.ca/~jnutini/documents/mlrg_ParallelCD.pdf) et [l](http://josephsalmon.eu/enseignement/TELECOM/MDI720/CD_fr.pdf)

Voir [ce rsum](https://tminka.github.io/papers/logreg/minka-logreg.pdf) pour une comparaison numrique de diffrents algorithmes.

Ce qu'il faut retenir est que `sklearn` dispose de plusieurs algorithmes qui ne sont pas quivalents en terme de capacit (rgularisation LASSO) ou de performance. Enfin, certains algorithmes ont t naturellement tendu au cas multi-class quand d'autres non.

## Stratgie de rglage des hyperparamtres

En dfinitive , il y a essentiellement $2$ (ventuellement $3$ avec l'_Elastic Net_) hyperparamtres  rgler lors de l'apprentissage :


*   Degr polynomial de la rgression
*   Paramtre de rgularisation $C$

En pratique, il est rare de devoir augmenter le degr polynomial mais en toute rigueur rien ne l'empche. Toutefois on limite ce degr car on a souvent affaire  des problmes de grande dimension d'entres ($n_{features}$). Dans de tels cas, augmenter le degr augmente le nombre de paramtres considrablement et complique la convergence des algorithmes d'optimisation. A l'inverse, un changement de $C$ n'a pas un impact significatif sur le temps de convergence. Au final, il faut tout de mme rgler le paramtre $C$.

On ne va pas construire de base de test ou de validation dans ce cas prcis, car le but n'est pas de construire un modle prdictif mais de trouver un modle qui explique les donnes. Toutefois pour viter le surapprentissage on va utiliser le leave-one out piour dterminer le meilleur paramtre de rgularisation $C$.

````python
from sklearn.model_selection import LeaveOneOut
from sklearn.linear_model import LogisticRegression
loo = LeaveOneOut()
loo.get_n_splits(X)
````

Pour chaque valeur de $C$, on va donc construire $24$ modles et on calculera la valeur moyenne de la prcision sur ces $24$ modles. Typiquement, pour $C=10$, cela donne :

````python
C = 10
res = []
for train_index, test_index in loo.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    clf = LogisticRegression(C=C)
    clf.fit(X_train.reshape(-1,1), y_train)
    res.append(clf.score(X_test.reshape(-1, 1), y_test))
````

On va crer une fonction qui  $C$ renvoie la valeur de leave-one-out

````python
def leave(C=10,loo=loo):
  res = []
  for train_index, test_index in loo.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    clf = LogisticRegression(C=C)
    clf.fit(X_train.reshape(-1,1), y_train)
    res.append(clf.score(X_test.reshape(-1, 1), y_test))
  return np.mean(np.array(res))
````

On trace ensuite les valuers de leave-one-out pour des $C$ qui varient entre $10^{-5}$ et $10^{3}$. Dans la mesure o la bonne gamme de valeurs pour $C$ n'est jamais connu  l'avance, il est souvent conseill d'chantillonner en log afin de couvrir un grand domaine de valeur possible.

````python
C_int = np.logspace(-5.,3.,100)
res_loo = []
for C in C_int:
  res_loo.append(leave(C,loo))
print(res_loo)
````

````python
configure_plotly_browser_state()
trace = Scatter(
    x=C_int,
    y=res_loo,
    mode='markers',
    marker=dict(size=13),
    text=[f"C : {C} " for C in C_int],
    textposition="top center",
    name="LOO",
)
fig = dict(data=[trace], layout={})
fig['layout']['xaxis'] = dict(title='C ridge regularization parameter')
fig['layout']['yaxis'] = dict(title='Leave One Out Accuracy')
fig['layout']['xaxis'] = dict(type="log")

iplot(fig)
````

Dans le graphe prcdent, on voit que les bonnes valeurs de $C$ se trouvent entre 0.01 et 0.1. On prend arbitrairement $C=0.02$ dans la suite et on entrane le modle final sur tous les points.
